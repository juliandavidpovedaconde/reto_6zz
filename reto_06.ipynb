Imprimir el factorial de un número natural n dado.

```python
x : int
x = int(input("ingrese un numero n: "))
if x<=0:
  print("porfavor ingrese un numero natural")
else:
  factorial = 1
  while x>=1:
   factorial=factorial*x
   x -= 1
print("El factorial es:", factorial)
```
